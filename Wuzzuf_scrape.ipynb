## Install libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
```

### Scrape data from Wuzzuf.com

```python
i=0
extracted_data = []
skills = []
while i<=100:
  url= f"https://wuzzuf.net/search/jobs/?a=navbl&filters%5Bcountry%5D%5B0%5D=Egypt&q=&start={i}"
  site = requests.get(url)
  soup = BeautifulSoup(site.content,'lxml')
  # Find job Title
  job_title = soup.find_all("h2",{'class':"css-m604qf"})
  # Find company name
  company_name = soup.find_all('a', {'class': 'css-17s97q8'})

  # Find location
  location = soup.find_all('span', {'class': 'css-5wys0k'})


  # Find the time posted
  time_posted = soup.find_all('div', {'class': 'css-4c4ojb'})


  # Find job type
  job_type = soup.find_all('span', {'class': 'css-1ve4b75 eoyjyou0'})

  i+=1
  time.sleep(5)

  for title, company, loc, time_p, jt in zip(job_title, company_name, location, time_posted, job_type):
      job_details = {
          'Job_Title': title.text,
          'Company_Name': company.text,
          'Location': loc.text,
          'Time_Posted': time_p.text,
          'Job_Type': jt.text,
      }
      extracted_data.append(job_details)
```

## Put the data into Data-frame

```python
jobs_df = pd.DataFrame(extracted_data)
```

```python
jobs_df
```

## Data manipulation and cleaning

```python
jobs_df.columns=jobs_df.columns.str.lower()
```

```python
jobs_df['state']=jobs_df['location'].str.split(',').str[-2]
jobs_df['city']=jobs_df['location'].str.split(',').str[-3]
```

```python
jobs_df['company_name']=jobs_df['company_name'].str.replace('-','')
```

```python
jobs_df
```

```python
def convert_to_today(time_str):
    if 'minute' in time_str or 'hour' in time_str or 'just now' in time_str:
        return 'Today'
    else:
        return time_str
jobs_df['time_posted'] = jobs_df['time_posted'].apply(lambda x: convert_to_today(x))
```
