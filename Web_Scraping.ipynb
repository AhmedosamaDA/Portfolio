## **Install libraries**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib
```

## **Loading data**

```python
url="https://web.greatervalleychamber.com/search"
```

```python
site = requests.get(url)
soup=BeautifulSoup(site.content,'lxml')
soup
```

## **Get categories**

```python
all_category=[]
get_category =soup.find_all("select",{'id':"content1_Directory1_DirectorySearch1_SearchCategory_ddKeyword"})
for category in get_category:
    options=category.find_all("option")
    for option in options:
        all_category.append(option.text)
```

```python
all_category=all_category[1:]
all_category
category=[]
for item in all_category:
    new=item.replace(' ','-')
    category.append(new)
```

```python
len(category)
```

## **Fetch necessary data from the 113 options saving in Data Frame**

```python
all_data = []
for name in category:
    url = f"https://web.greatervalleychamber.com/{name}"
    site = requests.get(url)
    print(name)
    print("Successful data extraction")
    soup = BeautifulSoup(site.content, 'lxml')
    b_name = soup.find_all("div", class_="ListingResults_All_ENTRYTITLELEFTBOX")
    b_info = soup.find_all("div", class_="ListingResults_Level1_MAINLEFTBOX")
    b_links = soup.find_all("div", "ListingResults_Level1_SOCIALMEDIA")

    for name, info, link in zip(b_name, b_info, b_links):
        get_a = link.find_all("a")
        links = []
        for a in get_a:
            unclean_link = a.get("href")
            extracted_url_part = unclean_link.split('URL=')[1].split("&")[0]
            url = urllib.parse.unquote(extracted_url_part)
            links.append(url)
        data = {"name": name.text.strip(), "info": info.text.strip(), "social": links}
        all_data.append(data)
```

```python
df= pd.DataFrame(all_data)
```

```python
df
```

## **Data Manipulation Using Pandas**

```python
df['location']=df["info"].str.split(',').str[0]
df['phone']=df['info'].str.split().str[-1]
df['postal_code']=df['info'].str.split().str[-2]
```
